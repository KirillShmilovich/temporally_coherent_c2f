```python
from cvae_test import cVAE
import pytorch_lightning as pl
from pl_callbacks import CheckpointEveryNSteps
from argparse import Namespace
```

```python
pl.seed_everything(42)
```

```python
hparams = pl.core.saving.load_hparams_from_tags_csv('ADP_meta_tags.csv')
```

```python
hparams['path'] = '../midway3_c2f/ADP_data/'
hparams['num_workers'] = 8
```

```python
model = cVAE(**hparams, mol_name='ADP', tops_fname=None)
```

```python
args = Namespace(**hparams)
logger = pl.loggers.TensorBoardLogger('./')
trainer = pl.Trainer.from_argparse_args(
        args,
        callbacks=[CheckpointEveryNSteps(2000)],
        val_check_interval=0.25,
        gpus=1,
        terminate_on_nan=True,
        gradient_clip_val=0.001,
        logger=logger,
    )
```

```python
trainer.fit(model)
```
